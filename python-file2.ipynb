# str.find(sub[, start[, end]]) and str.index(sub[, start[, end]])

- find(): Searches for the first occurences of substring sub within the string for the specified slice. Returns the starting index if found, and -1 if not found.

- index(): Same as find(), but raises a Valueerror if the substring is not found. use find() is you expect the substring might be missing and want to handle it gracefully(-1). Use index() if you expect the substring to be present and want an error if it's not.

```python
course_slogan = "Learn Data Engineering with Real-Time Projects"
tech_stack = "Azure DevOps and Python"
```

```python
data_pose = course_slogan.find("Data")
data_pose
```

```python
python_pose = tech_stack.find("Python")
print(python_pose)
```

```python
java_pose = course_slogan.find("java")
print(java_pose)
```

```python
java_pose = course_slogan.index("java")
print(java_pose)
```

## list.extend(iterable)                        ## list is an iterable

- Adds all the items from an iterable(like another list, tuple or string) to the end of the current list.
- Modifies the list in-place.        (original list is changed)
- Differs from append which adds the iterable as a single element.

```python
data_science_courses = ["DS Intro", "Python for DS"]
data_eng_courses = ["DE Intro", "SQL for DE", "ADF"]
new_tools = ("Docker", "Kubernetes")
```

```python
data_science_courses.extend(data_eng_courses)   #"DS Intro", "Python for DS", "DE Intro", "SQL for DE", "ADF"
data_science_courses
```

```python
batch_1_students = ["Alice", "Bob"]
batch_2_students = ["Charlie", "David"]
batch_1_students.extend(batch_2_students)
batch_1_students
```

```python
letters = ["a", "b"]
letters.extend("cde")
letters
```

```python
empty_list = []
empty_list.append([1,2,3])
empty_list
```

```python
empty_list = []
empty_list.extend([1,2,3])
empty_list
```
