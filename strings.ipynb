## More Example's on Strings

```python
name = "Suresh"
town = "Hindupur"
course = "Data Science"
```

```python
#1 Name in lower case
lower_case = name.lower()
print(f"1. lower_case: {lower_case}")
```

```python
#2 Town in upper case
upper_case = town.upper()
print(f"2. upper_case: {upper_case}")
```

```python
#3 Course in Lower Case
lower_case_course = course.lower()
print(f"3. lower_case_course: {lower_case_course}")
```

```python
user_input = "data science"
if user_input.lower() == course.lower():
    print("4. User input matches the course name. (case-insensitive).")
```

```python
# 5. Converting mixed case to uppercase
mixed_case2 = "PYthoN PROgrammING"
upper_mixed2 = mixed_case2.upper()
print(f"5. Uppercase mixed case: '{upper_mixed2}'")
```

```python
# 6. Converting mixed case to lowercase
lower_mixed2 = mixed_case2.lower()
print(f"6. Lowercase mixed case: '{lower_mixed2}'")
```

```python
# 7. Applying to a string with numbers/symbols (they remain unchanged)
roll_number = "Batch-11262-CM-038"
upper_case = roll_number.upper()
lower_case = roll_number.lower()
print(f"7. Uppercase roll number: '{upper_case}', Lowercase roll number: '{lower_case}'")
```

```python
# 8. Chaining after concatenation
combine = name + " from " + town + " is learning " + course + "at Edukron."
upper_combine = combine.upper()
print(f"8. Uppercase combined string: '{upper_combine}'")
```

```python
# 9. Applying to an already lowercase string
already_lower2 = "hello world"
output = already_lower2.lower()
print(f"9. Applying lower() to already lowercase string: '{output}'")
```

```python
# 10. Applying to an already uppercase string
already_upper2 = "HELLO WORLD"
output = already_upper2.upper()
print(f"10. Applying upper() to already uppercase string: '{output}'")
```

```python
## Concatenation (+) and Repetition (*)
company = "Edukron"
job_title = "Senior Data Analyst"
```

```python
# 1. Combining institute and focus area
full_title = "I'm working in " + company + " as a " + job_title
print(f"Full title: {full_title}")
```

```python
# 2. Creating a course welcome message
wishes = "All The Very Best"
welcome_message = "Welcome to " + company + "'s company and Wishing you " + wishes + "!"
print(f"Welcome message: {welcome_message}")
```

```python
# 3. Repeating a separator
repititor = "+*@&% " * 10
print(f"{repititor}")
repititor2 = "&%$ " * 5
print(f"{repititor2}" + " SURESH " + (repititor2) + " HINDUPUR " + (repititor2))
print(f"output:\n{repititor2}\n{repititor2}\n{repititor2}")
```

```python
# 4. Emphasizing a keyword
course = "Data Science @ "
repititor3 = course * 3
print(f"Emphasize the course = + {repititor3}")
```

```python
# 5. Building a file path (simple example, os.path.join is better)
path = "/courses/"
location_id = "PY101DS"
location_path = path + location_id
print(f"location Path: + {location_path}")
```

```python
city = "Banglore"
underline_city = "=" * len(city)
print(f"City:\n{city}\n{underline_city}")
```
