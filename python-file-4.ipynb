## Tuple

- Tuples are immutable(un-changeable), ordered sequence of item.
- They are defined using parentheses (), with items separated by commas.
- Like lists they can contain items of diff data types.
- Becuase they are immutablem they are often used for fixed collections of item. (Day-7 13:01)

- tuple support indexing, slicing, len() and membership testing(inm not in) just like lists.
- they lack methods that modify the tuple (like append, remove, sort, etc)
- have only two primary methods: Count() and Index()

## tuple.count(item)

- return the number of time the specified item appears in the tuple

```python
tech_skills = ["Python", "SQL", "Git", "CI/CD", "python", "SQL", "Python"]
rating = (5,4,5,5,3,4,5)
```

```python
tech_skills.count("Python")
```

```python
tech_skills.count("SQL")
```

```python
rating.count(5)
```

```python
rating.count(4)
```

```python
rating = (5,4,5,5,3,4,5)
rating_5_count = rating.count(5)
rating_5_count
```

```python
# count in a tuple with mixed types

mixed_tuple = (1,"Python", True, 1, "Python", 1, False)  # True = 1, False = 0
```

```python
mixed_tuple.count(1)
```

```python
nested = (1,(2,3),4,(2,3))
nested.count((2,3))
```

## tuple.index(item, [, start [end]])

- Return the index of the first occurence of the specified item.
- Raises a ValueError if the item is not found.
- Option start and end argument limit the search to a slice.

```python
tech_skills = ["Python", "SQL", "Git", "CI/CD", "Python", "SQL"]
rating = (5,4,5,5,3,4,5)
```

```python
sql_index = tech_skills.index("SQL")
sql_index
```

```python
python_index = tech_skills.index("Python")
python_index
```

```python
git_index = tech_skills.index("Git")
git_index
```

```python
java_index = tech_skills.index("Java")
java_index
```

```python
rating = (5,4,5,5,3,4,5)
rating_3_index = rating.index(3)
rating_3_index
```

```python
# find "Python" starting the search from the index 1
tech_skills = ["Python", "SQL", "Git", "CI/CD", "Python", "SQL", "Python"]
python_index_from_1 = tech_skills.index("Python", 5)
python_index_from_1
```

```python
sql_index_from_1 = tech_skills.index("SQL",2)
sql_index_from_1
```
