# list.pop([index])

- Remove and return the item at the given index.
- If index is omitted , it removes and return the last item.
- Raises and indexError if the list is empty or the index is out of range.
- Modifies the list in-place.

```python
course_list = ["Python Basics", "Data Science", "Data Engineering"]
tasks = ["Review Code", "Update Docs", "Deploy"]
```

```python
# Pop the last item

last_course = course_list.pop()
last_course
```

```python
course_list
```

```python
first_course = course_list.pop(0)
first_course
```

```python
last_task = tasks.pop()
last_task
```

```python
tasks
```

```python
last_task = tasks.pop()
last_task
```

```python
numbers = [10,20,30,40]
second_last = numbers.pop(-2)
second_last
```

## list.sort(key = None, reverse = False)

- Sorts the item of the list in-place
- key : A function to be called on each element before making comparision(key=str.lower) for cas-insentive sort.
- reverse = True, Sorts in descending order.

```python
course_list = ["Data Engineering", "Azure DevOps", "Data Science", "Python Basics"]
student_counts = [350, 420, 380, 500, 450]
mixed_case_counts = ["azure DevOps", "Python Basics", "Data Sceince"]
```

```python
#sort course list alphabetically

course_list.sort()
course_list
```

```python
student_counts.sort()
student_counts
```

```python
mixed_case_counts.sort()
mixed_case_counts
```

```python
mixed_case_counts.insert(3,"basic Python")
mixed_case_counts
```

```python
print(mixed_case_counts)
mixed_case_counts.sort(key=str.lower)
mixed_case_counts
```

```python
mixed_case_counts.sort(key=str.lower, reverse=True)
mixed_case_counts
```

## Membership Testing (in, not in)
- Checks if an item exists within th list.
- Returns True or False                  (# Day7 5:00 min)

```python
course_list = ["Data Science", "Data Engineering", "Azure DevOps"]
tech_skills = ["Python", "SQL", "Git", "CI/CD"]
```

```python
# check if "Data Sceince" is in th course list
is_ds_offered = "Data Science" in course_list
is_ds_offered
```

```python
is_ml_offered = "Machine Learning" in course_list
is_ml_offered
```

```python
is_python_not_in_course_list = "Python" not in course_list
is_python_not_in_course_list
```

```python
is_ds_not_in_course_list = "Data Science" not in course_list
is_ds_not_in_course_list
```

```python
is_python_not_in_tech_skills = "Python" not in tech_skills
is_python_not_in_tech_skills
```
